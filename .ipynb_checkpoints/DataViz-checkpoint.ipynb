```python
#Read in Fitbit data
iFile = '~/Downloads/FitbitFebMarch2016.csv'
data = pd.read_csv(iFile)

#Print column headers
#print data.columns.values

def GetFormattedDate (rawDate, f):
    #Add padding if necessary   
    p = [i.rjust(2,'0') for i in rawDate.split('/')]
    
    #Add prefix to years (assuming all data from 21st century!)
    if (len(p[2]) == 2):
        p[2] = '20' + p[2]
    
    #Get formatted datetime object
    fdate = str(str(p[0] + '/' + p[1] + '/' + p[2]))
    
    if (f == 'num'):
        fdate = date2num(datetime.datetime.strptime(fdate,"%m/%d/%Y"))
    elif (f == 'obj'):
        fdate = datetime.datetime.strptime(fdate, "%m/%d/%Y")
    elif (f == 'ymd'):
        fdate = datetime.datetime.strptime(fdate, "%Y-%m-%d")
    return fdate

#d = GetFormattedDate('2/18/16')
#x = datetime.date.today()
#print d.strftime("%m %d %Y")
#print d.timetuple()
```

```python
#Misc. utility functions for dates with matplotlib: plt.matplotlib.dates.

from matplotlib.dates import date2num, num2date

subset = data[['Date', 'MinutesAsleep']]
tuples = [tuple(x) for x in subset.values]

x = [GetFormattedDate(date, "num") for (date, value) in tuples]
y = [value for (date, value) in tuples]

print(x)
print(y)

fig = plt.figure()

graph = fig.add_subplot(111)

# Plot the data as a red line with round markers
graph.plot(x,y)

# Set the xtick locations to correspond to just the dates you entered.
graph.set_xticks(x)

# Set the xtick labels to correspond to just the dates you entered.
graph.set_xticklabels(
        [datetime.date.strftime(num2date(date), "%d") for date in x]
        )

plt.show()
```
